# Updating Rows and Columns - Modifying Data Within DataFrames


```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}

df = pd.DataFrame(people)

df.head()
```

```python
df.columns
```

```python
# name for all of the columns
df.columns = ['first name', 'last name', 'email']
```

```python
df.head()
```

```python
df.columns = [x.upper() for x in df.columns]
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
df.rename(columns={'FIRST_NAME':'first','LAST_NAME':'last'}, \
          inplace=True)
```

```python
df.head()
```

```python
df.loc[2] = ['John','Smith','JohnSmith@email.com']
```

```python
df.head()
```

```python
df.loc[2,['last','EMAIL']] = ['Doe','JohnDoe@emial.com']
```

```python
df.head()
```

```python
df.loc[2,'last'] = 'Smith'
```

```python
df.head()
```

```python
# at is as same as loc

df.at[2, 'last'] = 'Doe'
```

```python
df.head()
```

```python
filt = (df['EMAIL'] == 'JohnDoe@emial.com')
df[filt]['last'] = 'Smith'
```

```python
df.head()
```

```python
df.loc[filt,'last'] = 'Smith'
```

```python
df.head()
```

```python
# how to update multiple rows of data

df['EMAIL'] = df['EMAIL'].str.lower()
```

```python
df.head()
```

```python
# apply
# map
# applymap
# replace 
```

```python
# apply on dataFrame and Series

df['EMAIL'].apply(len)
```

```python
def update_email(email):
    return email.upper()
```

```python
df['EMAIL'].apply(update_email)
```

```python
df['EMAIL'] = df['EMAIL'].apply(update_email)
```

```python
df.head()
```

```python
func = lambda x: x.lower()
```

```python
df['EMAIL'] = df['EMAIL'].apply(func)
```

```python
df.head()
```

```python
df.apply(len)
```

```python
len(df['EMAIL'])
```

```python
df.apply(len,axis='columns') # axis=1
```

```python
df.apply(pd.Series.min)
```

```python
df.apply(lambda x: x.min())
```

```python
df.applymap(len)
```

```python
df.applymap(str.lower)
```

```python
df['first'].map({'Corey':'Chris','Jane':'Mary'})
```

```python
df['first'].replace({'Corey':'Chris','Jane':'Mary'})
```
