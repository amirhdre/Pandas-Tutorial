# Updating Rows and Columns - Modifying Data Within DataFrames


```python
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"],
}

df = pd.DataFrame(people)

df.head()
```

```python
df.columns
```

```python
# set new names for all of the columns
df.columns = ["first name", "last name", "email"]
```

```python
df.head()
```

```python
# capitalize all of the columns labels
df.columns = [x.upper() for x in df.columns]
```

```python
df.head()
```

```python
# replace a string in column labels
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.head()
```

```python
# rename only some of the column names (not all at once like above)
df.rename(columns={"FIRST_NAME": "first", "LAST_NAME": "last"}, inplace=True)
```

```python
df.head()
```

```python
# Access a group of rows and columns by label(s) or a boolean array and then change their contents
df.loc[2] = ["John", "Smith", "JohnSmith@email.com"]
```

```python
df.head()
```

```python
df.loc[2, ["last", "EMAIL"]] = ["Doe", "JohnDoe@emial.com"]
```

```python
df.head()
```

```python
df.loc[2, "last"] = "Smith"
```

```python
df.head()
```

```python
# at is as same as loc
df.at[2, "last"] = "Doe"
```

```python
df.head()
```

```python
filt = df["EMAIL"] == "JohnDoe@emial.com"
df[filt]["last"] = "Smith"
```

```python
df.head()
```

```python
df.loc[filt, "last"] = "Smith"
```

```python
df.head()
```

```python
# how to update multiple rows of data at once
df["EMAIL"] = df["EMAIL"].str.lower()
```

```python
df.head()
```

```python
# apply
# map
# applymap
# replace
```

```python
## apply Series
# Invoke function on values of Series
df["EMAIL"].apply(len)
```

```python
## apply Dataframe
# Invoke function on values of Dataframe

df.apply(len)
```

```python
# define a function to invoke on values of the Series
def update_email(email):
    return email.upper()

# invoke update_email function on values of Series
df["EMAIL"].apply(update_email)
```

```python
# apply update_email function on values of Series
df["EMAIL"] = df["EMAIL"].apply(update_email)
```

```python
df.head()
```

```python
# define a function to invoke on values of the Series
func = lambda x: x.lower()

# apply func function on values of Series
df["EMAIL"] = df["EMAIL"].apply(func)
```

```python
df.head()
```

```python
df.apply(len)
```

```python
len(df["EMAIL"])
```

```python
df.apply(len, axis="columns")  # axis=1
```

```python
df.apply(pd.Series.min)
```

```python
df.apply(lambda x: x.min())
```

```python
# invoke a function on values of dataframe
df.applymap(len)
```

```python
df.applymap(str.lower)
```

```python
df["first"].map({"Corey": "Chris", "Jane": "Mary"})
```

```python
df["first"].replace({"Corey": "Chris", "Jane": "Mary"})
```

```python
df["first"] = df["first"].replace({"Corey": "Chris", "Jane": "Mary"})
```
