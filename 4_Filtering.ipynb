# Filtering - Using Conditionals to Filter Rows and Columns

```python
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"],
}

df = pd.DataFrame(people)

df.head()
```

```python
# create a filter 
filt = df["last"] == "Doe"
```

```python
# filter df 
df[filt]
```

```python
# filter df in-place
df[df["last"] == "Doe"]
```

```python
# Access a group of rows and columns by label - while having a filter
df.loc[filt, "email"]
```

```python
# design a filter with two conditions 
filt = (df["last"] == "Doe") & (df["first"] == "John")

# use filter on dataframe
df.loc[filt]
```

```python
# not a filter
df.loc[~filt]
```

# Run filters on survey data

```python
import pandas as pd

df = pd.read_csv("data/survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("data/survey_results_schema.csv", index_col="Column")

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)
```

```python
df.head(2)
```

```python
# desing filter
filt = df["ConvertedComp"] > 70_000

# use filter and access three columns of filtered dataframe
df.loc[filt, ["Country", "LanguageWorkedWith", "ConvertedComp"]]
```

```python
# design a filter
countries = ["United States", "India", "United Kingdom", "Germany", "Canada"]
filt2 = df["Country"].isin(countries)

# use the filter
df.loc[filt2, ["Country", "LanguageWorkedWith", "ConvertedComp"]]
```

```python
# design a filter
filt3 = df["LanguageWorkedWith"].str.contains("Python", na=False)

# use the filter
df.loc[filt3, "LanguageWorkedWith"]
```
