```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["Corey", "Jane", "John", "Chris", np.nan, None, "NA"],
    "last": ["Schafer", "Doe", "Doe", "Schafer", np.nan, np.nan, "Missing"],
    "email": [
        "Corey∏MSchafer@gmail.com",
        "JaneDoe@email.com",
        "JohnDoe@email.com",
        None,
        np.nan,
        "Anonymous@email.com",
        "NA",
    ],
    "age": ["33", "55", "63", "36", None, None, "Missing"],
}

df = pd.DataFrame(people)
```

```python
df
```

```python
# Replace values of the DataFrame with other values dynamically
df.replace("NA", np.nan, inplace=True)
df.replace("Missing", np.nan, inplace=True)

df
```

```python
df.dropna()
```

```python
df.dropna(axis="index", how="any")  # default parameters
```

```python
df.dropna(axis="index", how="all")
# all means that all of the values should be missing in oder to drop that row
```

```python
df.dropna(axis="columns", how="all")
```

```python
df.dropna(axis="columns", how="any")
```

```python
df.dropna(axis="index", subset=["email"])
```

```python
df.dropna(axis="index", how="all", subset=["last", "email"])
```

```python
df.dropna(axis="index", how="any", subset=["last", "email"])
```
